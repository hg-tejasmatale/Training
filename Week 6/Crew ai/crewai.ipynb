```python
import os
from dotenv import load_dotenv
from crewai import Agent, Task, Crew, LLM

load_dotenv()

gemini_llm = LLM(
    model="gemini/gemini-2.5-flash",
    api_key=os.getenv("GEMINI_API_KEY"),
    temperature=0.7
)

# Update your researcher agent to use the new LLM
researcher = Agent(
    role="AI Researcher",
    goal="Gather the latest AI trends from 2024",
    backstory="You are an expert researcher who writes reports on emerging AI trends.",
    verbose=True,
    llm=gemini_llm  # Pass the Gemini LLM here
)

# The rest of your task and crew code remains the same
research_task = Task(
    description="Identify the top 3 AI trends in 2024 using online sources.",
    expected_output="A list of the top 3 trends with a short summary of each.",
    agent=researcher
)

crew = Crew(
    agents=[researcher],
    tasks=[research_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

```python
writer = Agent(
    role = "You are an blog wirter",
    goal = "Write the blog for the AI enthuist baed on the context",
    backstory= "You are and expert blog writer who writes blogs on the given context professionaly",
    llm = gemini_llm
)

write_task = Task(
    description = "Write a blog post based on AI findings.",
    expected_output="A 300-word post",
    agent=writer,
    context=[research_task]
)
```

```python
crew = Crew(
    agents=[writer],
    tasks=[write_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

```python
import fitz  # PyMuPDF
from crewai.tools import tool
from crewai import Agent, Task, Crew

# Define a custom tool to read PDF text directly
@tool("PDFTextReader")
def pdf_text_reader(pdf_path: str):
    """Extracts all text from a PDF file."""
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

# Create the agent with the custom tool
researcher_pdf = Agent(
    role="Academic Researcher",
    goal="Explain PDF content clearly for students",
    backstory="You are expert at simplifying complex documents.",
    tools=[pdf_text_reader], # Use your new custom tool
    verbose=True,
    llm=gemini_llm
)

pdf_task = Task(
    description="Use the PDFTextReader tool to read '2024-wttc-introduction-to-ai.pdf' and write an explanation for students.",
    expected_output="A 300-word educational post.",
    agent=researcher_pdf,
)

crew = Crew(agents=[researcher_pdf], tasks=[pdf_task], verbose=True)
result = crew.kickoff()
print(result)
```
