```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('retail_churn_data.csv')
df.head()
```

```python
LabelEncoder = LabelEncoder()
df['Churn'] = LabelEncoder.fit_transform(df['Churn'])
df.head()
```

```python
x = df.drop(['Customer_ID', 'Churn'], axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
base_tree = DecisionTreeClassifier()
```

```python
bagging_model = BaggingClassifier(
    estimator=base_tree,
    n_estimators=100,
    bootstrap=True,
    max_samples=0.9,
    bootstrap_features=True,
    random_state=42
)

bagging_model.fit(X_train, y_train)
y_pred = bagging_model.predict(X_test)

print("Bagging Model Accuracy:", accuracy_score(y_test, y_pred))
```

Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df_rand = pd.read_csv("healthcare_readmission_500.csv")
df_rand.head()
```

```python
x = df_rand.drop(['Patient_ID', 'Readmitted'], axis=1)
y = df_rand['Readmitted']
```

```python
X_train, X_test , y_train, y_test = train_test_split(
    x,y,
    test_size=0.28,
    random_state=666
)
```

```python
model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    max_features = "log2",
    bootstrap= True,
    random_state=666,
)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
